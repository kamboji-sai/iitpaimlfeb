```python
score = int(input("Enter your score"))

if score >= 90:
    print("Grade: A")
    print("Excellent work!")

if score >= 80 and score < 90:
    print("Grade: B")
    print("Good job!") 

if score >= 70 and score < 80:
    print("Grade: C")
    
    print("Satisfactory")

if score >= 60 and score < 70:
    print("Grade: D")
    print("Needs improvement")

if score < 60:
    print("Grade: F")
    print("Failed")    

if score == 100:
    print("Perfect score")

if score < 50:
    print("Please see instructor")
```
